```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
df = pd.read_csv("C:/Users/ASUS X/Desktop/fashion 2024/January - March/evolution_of_sales_and_quantity_change - Jan_vs_Mar_2024.csv")
```

```python
#Preview first 5 rows
print(df.head())
```

```python
# Set up the plot
fig, ax1 = plt.subplots(figsize=(12, 6))
bar_width = 0.35
index = range(len(df))

# Bar charts for January and March sales
bars_jan = ax1.bar([i - bar_width/2 for i in index], df['jan_sales'], bar_width, label='January Sales', color='steelblue')
bars_mar = ax1.bar([i + bar_width/2 for i in index], df['mar_sales'], bar_width, label='March Sales', color='orange')

# Labels and ticks
ax1.set_ylabel('Sales (RON)', fontsize=10)
ax1.set_title('Sales Evolution: January vs March 2024 – by Product', fontsize=12)
ax1.set_xticks(index)
ax1.set_xticklabels(df['full_product_name'], rotation=45, ha='right')
ax1.legend(loc='upper left')

# Create second Y-axis for percentage change
ax2 = ax1.twinx()
ax2.plot(index, df['sales_change_pct'], color='darkred', marker='o', label='Sales Change (%)')
ax2.set_ylabel('Sales Change (%)', color='darkred', fontsize=10)
ax2.tick_params(axis='y', labelcolor='darkred')

# Add text labels above bars
for i, (jan, mar) in enumerate(zip(df['jan_sales'], df['mar_sales'])):
    ax1.text(i - bar_width/2, jan + 500, f"{int(jan)}", ha='center', fontsize=8)
    ax1.text(i + bar_width/2, mar + 500, f"{int(mar)}", ha='center', fontsize=8)

# Add text labels for percentage change
for i, pct in enumerate(df['sales_change_pct']):
    ax2.text(i, pct + 0.3, f"{pct:.0f}%", ha='center', fontsize=8, color='darkred')

# Save and show plot
plt.tight_layout()
plt.savefig("charts/sales_evolution_jan_vs_mar.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Sales Evolution: January vs March 2024 - by Product
# Bar chart for sales + line chart for %

```python
#Load dataset
df = pd.read_csv("C:/Users/ASUS X/Desktop/fashion 2024/January - March/product_performance_comparison_Jan_vs_Mar_2024.csv")
```

```python
#Print first 5 rows
print(df.head())
```

```python
#Create figure and axes
fig, ax1 = plt.subplots(figsize=(12, 6))
bar_width = 0.35
index = range(len(df))

#Bar chart: January & March
bars1 = ax1.bar([i - bar_width/2 for i in index], df['jan_sales'], bar_width, label='January Sales', color='skyblue')
bars2 = ax1.bar([i + bar_width/2 for i in index], df['mar_sales'], bar_width, label='March Sales', color='orange')

#Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height + 500, f'{int(height):,}', ha='center', fontsize=8)
for bar in bars2:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height + 500, f'{int(height):,}', ha='center', fontsize=8)

#Line chart for percentage change
ax2 = ax1.twinx()
ax2.plot(index, df['sales_change_pct'], color='darkred', marker='o', label='Sales Change %')

#Add % labels on line chart
for i, value in enumerate(df['sales_change_pct']):
    ax2.text(i, value + 50, f'{value:.0f}%', ha='center', fontsize=8, color='darkred')

#Customize axes
ax1.set_xlabel('Product')
ax1.set_ylabel('Sales (RON)')
ax2.set_ylabel('Sales Change %')
ax1.set_title('Sales Evolution January vs March – by Product')
ax1.set_xticks(index)
ax1.set_xticklabels(df['product'], rotation=45, ha='right')

#Combine legends
lines_labels = [ax.get_legend_handles_labels() for ax in [ax1, ax2]]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
ax1.legend(lines, labels, loc='upper left')

#Save and show
plt.tight_layout()
plt.savefig("charts/sales_evolution_jan_vs_mar.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Top supplier by average discount January and March 2024

```python
#Load dataset
df = pd.read_csv("C:/Users/ASUS X/Desktop/fashion 2024/January - March/top_supplier_by_avg_discount_Jan_vs_Mar_2024.csv")
```

```python
#Print first 5 rows
print(df.head())
```

```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Plot
plt.figure(figsize=(14, 6))
barplot = sns.barplot(data=df, x='Supplier', y='avg_discount', hue='Month')

#Add data labels
for container in barplot.containers:
    barplot.bar_label(container, fmt='%.1f', fontsize=8, label_type='edge')

#Customize labels and title
plt.title('Top Supplier by Average Discount - Jan vs Mar')
plt.ylabel('Average Discount (%)')
plt.xlabel('Supplier')
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("charts/top_supplier_by_average_discount_jan_vs_mar.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
jan_discounts = df[(df['Month'] == 'January') & (df['avg_discount'] > 0)]
mar_discounts = df[(df['Month'] == 'March') & (df['avg_discount'] > 0)]
jan_suppliers = set(jan_discounts['Supplier'])
mar_suppliers = set(mar_discounts['Supplier'])
stopped_discounts = jan_suppliers - mar_suppliers
stopped_df = jan_discounts[jan_discounts['Supplier'].isin(stopped_discounts)]
print("Suppliers that stopped offering discounts in March:")
print(stopped_df[['Supplier', 'avg_discount']])
```
