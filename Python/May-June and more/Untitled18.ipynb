# Average discount by supplier May - June 2024

```python
#Load dataset
df = pd.read_csv("C:/Users/ASUS X/Desktop/fashion 2024/May-June/Avg_Discount by Supplier in May - June rank 5.csv")
```

```python
#Print first 5 rows
print(df.head())
```

```python
#Create pivot table
pivot_df = df.pivot(index='Supplier', columns='month_name', values='avg_discount')

#Sort by May values to show most discounted suppliers first
if 'May' in pivot_df.columns:
    pivot_df = pivot_df.sort_values(by='May', ascending=False)
pivot_df.plot(kind='bar', figsize=(10, 6))

#Customize chart
plt.title("Top 5 Suppliers by Avg. Discount % – May vs June")
plt.ylabel("Average Discount (%)")
plt.xlabel("Supplier")
plt.xticks(rotation=45)
plt.legend(title="Month")

#Add labels
for idx, row in pivot_df.iterrows():
    for i, val in enumerate(row):
        if not pd.isna(val):
            plt.text(x=list(pivot_df.index).index(idx) + (i - 0.5) * 0.2, y=val + 0.2, s=f"{val:.1f}%", ha='center', fontsize=9)

#Layout and show chart
plt.tight_layout()
plt.show()
```

# Distribution of transaction and sales by discount interval October and November 2024

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#Load dataset
df = pd.read_csv("C:/Users/ASUS X/Desktop/fashion 2024/October - November/Distribution of transaction and sales by discount interval Oct-Nov 2024.csv")
```

```python
#Print first 5 rows
print(df.head())
```

```python
#Order intervals
interval_order = ['0%', '0-5%', '5-10%', '10-20%', '20-30%', '>30%']
df['discount_interval'] = pd.Categorical(df['discount_interval'], categories=interval_order, ordered=True)

# Sort by range
df = df.sort_values('discount_interval')

# Pivot for number_of_transactions
pivot_tx = df.pivot_table(
    index='discount_interval',
    columns='month_name',
    values='number_of_transactions',
    fill_value=0
)

# Pivot for total_net_sales
pivot_sales = df.pivot_table(
    index='discount_interval',
    columns='month_name',
    values='total_net_sales',
    fill_value=0
)

# --------------------------
# Chart 1 – Transactions
# --------------------------

#Number of transaction by discount interval
ax1 = pivot_tx.plot(
    kind='bar',
    figsize=(9,5),
    color=['#FFC107', '#4CAF50']  
)

#Title and labels
plt.title('Number of Transactions by Discount Interval\nOctober vs. November 2024')
plt.xlabel('Discount Interval')
plt.ylabel('Number of Transactions')
plt.legend(title='Month')

# Add labels on bars
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.0f', padding=3)

#Layout and show chart
plt.tight_layout()
plt.show()

# --------------------------
# Chart 2 – Total Net Sales
# --------------------------

#Total net sales by discount interval 
ax2 = pivot_sales.plot(
    kind='bar',
    figsize=(9,5),
    color=['#FFC107', '#4CAF50']
)

#Title and labels
plt.title('Total Net Sales by Discount Interval\nOctober vs. November 2024')
plt.xlabel('Discount Interval')
plt.ylabel('Total Net Sales (RON)')
plt.legend(title='Month')

# Add labels on bars
for container in ax2.containers:
    ax2.bar_label(container, fmt='%.0f', padding=3)

#Layout and show chart
plt.tight_layout()
plt.show()
```

# Total net sales by product group October - November 2024

```python
#Load dataset
df = pd.read_csv("C:/Users/ASUS X/Desktop/fashion 2024/October - November/Total net sales by product group Oct-Nov 2024.csv")
```

```python
#Print first 5 rows
print(df.head())
```

```python
#Convert all month names to uppercase and remove spaces
df['month_name'] = df['month_name'].str.upper().str.strip()
df['Product_group'] = df['Product_group'].str.upper().str.strip()

# Select November and sort by top 3
november_data = df[df['month_name'] == 'NOVEMBER']
november_sorted = november_data.sort_values('total_net_sales', ascending=False)
top3_groups = november_sorted['Product_group'].head(3).tolist()

print("Top 3 Product Groups in November:", top3_groups)

# Filter Top 3
df_top3 = df[df['Product_group'].isin(top3_groups)]

# Pivot
pivot_top3 = df_top3.pivot_table(
    index='Product_group',
    columns='month_name',
    values='total_net_sales',
    fill_value=0
)

# Make sure values are numeric
pivot_top3 = pivot_top3.apply(pd.to_numeric, errors='coerce')

print(pivot_top3)

# Plot
ax = pivot_top3.plot(
    kind='bar',
    figsize=(12, 6),              
    width=0.6,                    
    color=['#FFC107', '#4CAF50']  
)

#Titles and labels
plt.title('Top 3 Product Groups - Total Net Sales\nOctober vs. November 2024', fontsize=14)
plt.xlabel('Product Group', fontsize=12)
plt.ylabel('Total Net Sales (RON)', fontsize=12)
plt.xticks(rotation=45, ha='right')  
plt.legend(title='Month')

# Value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

#Layout and show chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = df[['month_name', 'Supplier', 'avg_discount']]

# creează pivot table ca să ai May și June pe coloane separate
pivot_df = df.pivot_table(index='Supplier', columns='month_name', values='avg_discount').fillna(0)

# sortez după June ca să fie ordinea relevantă (opțional)
pivot_df = pivot_df.sort_values('June', ascending=False)

# extrage valorile
suppliers = pivot_df.index.tolist()
discount_may = pivot_df.get('May', pd.Series(0, index=pivot_df.index))
discount_june = pivot_df.get('June', pd.Series(0, index=pivot_df.index))

# pregătesc pozițiile barelor
x = np.arange(len(suppliers))
width = 0.35

# fac graficul
fig, ax = plt.subplots(figsize=(12,6))

bars1 = ax.bar(x - width/2, discount_may, width, label='Discount May', color='steelblue')
bars2 = ax.bar(x + width/2, discount_june, width, label='Discount June', color='darkorange')

# adaug labels deasupra barelor
for bar in bars1:
    height = bar.get_height()
    if height > 0:
        ax.annotate(f'{height:.2f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 5),
                    textcoords="offset points",
                    ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    if height > 0:
        ax.annotate(f'{height:.2f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 5),
                    textcoords="offset points",
                    ha='center', va='bottom')

ax.set_xlabel('Supplier')
ax.set_ylabel('Average Discount (%)')
ax.set_title('Average Discount by Supplier – May vs June 2023')
ax.set_xticks(x)
ax.set_xticklabels(suppliers, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
